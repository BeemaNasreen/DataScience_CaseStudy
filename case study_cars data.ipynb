# Bar graph of male vs female buyers participated in sales

```python
data['Buyer Gender'].value_counts().plot(kind='bar',width = 0.2,figsize=(7,5),color =['violet','purple']);
plt.xlabel("Gender")
plt.ylabel("Count of People")
plt.title("Count of male vs female buyers participated in the sales",fontsize=16);
```
